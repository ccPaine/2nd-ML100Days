```python

# 載入套件
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice

from sklearn.cluster import AgglomerativeClustering
```

```python
# 設定 2D 樣板資料
n_samples = 1500
random_state = 100

# 生成 雙半月 資料點
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)

# 生成 緊密三群 資料點
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)

# 生成 2維均勻分布 資料點
no_structure = np.random.rand(n_samples, 2)
```

# 作業
* 試著完成剩下三種 2D 樣板資料的繪圖, 並觀察有何不同

```python
sc = StandardScaler()
noisy_moons = sc.fit_transform(noisy_moons[0])
blobs = sc.fit_transform(blobs[0])
no_structure = sc.fit_transform(no_structure)
```

```python
n_clusters = 2
aggc_ward = AgglomerativeClustering(linkage = 'ward' , n_clusters = n_clusters)
aggc_complete = AgglomerativeClustering(linkage = 'complete' , n_clusters = n_clusters)
aggc_average = AgglomerativeClustering(linkage = 'average' , n_clusters = n_clusters)
aggc = [aggc_ward , aggc_complete , aggc_average]

plt.figure(figsize=(18, 6))
for i in range(0 , 3):
    y_pred = aggc[i].fit_predict(noisy_moons)
    plt.subplot(1 , 3 , i + 1)
    plt.scatter(noisy_moons[:, 0], noisy_moons[:, 1], c = y_pred)
    if i == 0:
        plt.title('ward')
    elif i == 1:
        plt.title('complete')
    else:
        plt.title('average')
```

```python

n_clusters = 3
aggc_ward = AgglomerativeClustering(linkage = 'ward' , n_clusters = n_clusters)
aggc_complete = AgglomerativeClustering(linkage = 'complete' , n_clusters = n_clusters)
aggc_average = AgglomerativeClustering(linkage = 'average' , n_clusters = n_clusters)
aggc = [aggc_ward , aggc_complete , aggc_average]
plt.figure(figsize=(18, 6))
for i in range(0 , 3):
    y_pred = aggc[i].fit_predict(blobs)
    plt.subplot(1 , 3 , i + 1)
    plt.scatter(blobs[:, 0], blobs[:, 1], c = y_pred)
    if i == 0:
        plt.title('ward')
    elif i == 1:
        plt.title('complete')
    else:
        plt.title('average')
```
